```python
import pandas as pd
from IPython.display import display

# 導入第一個CSV文件，不使用默認索引列
df1 = pd.read_csv('2023 大學熱門學群.csv', index_col=False)

# 導入第二個CSV文件，不使用默認索引列
df2 = pd.read_csv('2024 大學熱門學群.csv', index_col=False)

# 交集
intersection = pd.merge(df1, df2, how='inner')

# 聯集
union = pd.concat([df1, df2]).drop_duplicates()

# 差集
difference_df1_df2 = pd.concat([df1, df2, df2]).drop_duplicates(keep=False)
difference_df2_df1 = pd.concat([df2, df1, df1]).drop_duplicates(keep=False)

# 打印結果
print("分析近期兩年的大學熱門學群變化(資料從2023年到2024年)")
print('')
print("")
print("")
print("2023的七大熱門學群(資料參考大學問網站，排名依序由高到低，0到6分別為第1到第7名。)：\n")
display(df1)
print('')
print("2024七大熱門學群(資料參考大學問網站，排名依序由高到低，0到6分別為第1到第7名。)：\n")
display(df2)
print("")
print('')
print("交集結果："+"在這兩年間被大眾所公認的熱門學群：\n")
display(intersection)
print('')
print('')
print("聯集結果："+"在這兩年間曾經出現的熱門學群：")
print('')
display(union)
print('')
print('')
print("差集結果1："+"在2023年有出現的熱門學群而 2024 年卻沒有再出現的學群：")
print('')
display(difference_df1_df2)
print('')
print('')
print("差集結果2："+"在 2024 年取代 2023 年的熱門學群：")
print('')
display(difference_df2_df1)
print('')
print('')
print('')
print("分析結果：")
print('')
print("1.資訊學群會取代遊憩與運動學群，成為2024年的熱門學群，有一大理由是因為AI浪潮來襲，投入資工、資管系成為工程師對於許多人來說是不二選擇。因為當掌握了AI應用技術，等同於掌握到市場脈絡，在資訊方面供需依舊極大的情況下，成為工程師的薪資也是頗為可觀。而這或許也正是運動與遊憩學群被取代的原因。")
print("2.其餘六大學群依舊長期霸榜，僅有部分排名些微更動，有極大的可能性是因為，長久以來這六大學群在人們心中佔據極高的地位。像是社會與心理學群，之所以能在近兩年成為熱門學群，推估其背後原因，可能是因為現代人由於科技冷漠、人群疏離，未來心理醫生或是諮商師也會是一大熱門職業。")
print("3.資訊學群之所以沒有佔據榜首，有可能是因為即便是大AI時代，但真正對數理統計有興致的人還無法擴及到整個大環境。也就是資料科學興起，AI應用在各個領域成為趨勢，但是不見得所有人都熱衷於電腦科學資訊領域。")
```
